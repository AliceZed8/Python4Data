```python
%matplotlib inline  
import librosa
import librosa.display
import IPython
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

## Считываем аудио файл

```python
audio_data = 'test.mp3'
y, sr = librosa.load(audio_data)
```

Функция load считывает аудио-файл в виде временного рядя. Где, sr обозначает sample_rate.
Time series (временной ряд) представлен массивом.
sample_rate — это количество сэмплов на секунду аудио.
По умолчанию звук микшируется в моно. Но его можно передискретизировать во время загрузки до 22050 Гц. Это делается с помощью дополнительных параметров в функции librosa.load

## Анализ сигнала

### Отрисуем сигнал в амплитудно-временной форме

```python
print(type(y), type(sr))
print(y.shape, sr)
```

С помощью IPython.display можно проигрывать аудио в Jupyter Notebook, а с помощью display.waveshow формируются звуковые волны и мы можем посмотреть, как выглядит звук:

```python
import IPython.display as ipd
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr)
ipd.Audio(audio_data)
```

### Используя преобразование Фурье отрисовать частотный спектр

```python
# Compute the FFT of the waveform
from scipy.fft import fft
spectrum = fft(y)
 
# Frequency bins
frequencies = np.fft.fftfreq(len(spectrum), 1 / sr)
 
# Plot the frequency spectrum
plt.figure(figsize=(8, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(spectrum[:len(spectrum)//2]))
plt.xlabel('Частота (Hz)')
plt.ylabel('Амплитуда')
plt.title('Частотный спектр')
plt.grid(True)
plt.show()
```

### Cпектрограмма сигнала

Спектрограмма

Спектрограмма - это визуальный способ представления уровня или “громкости” сигнала во времени на различных частотах, 
присутствующих в форме волны. Обычно изображается в виде тепловой карты. .stft() преобразует данные в кратковременное преобразование 
Фурье. С помощью STFT можно определить амплитуду различных частот, воспроизводимых в данный момент времени аудиосигнала.

```python
X = librosa.stft(y)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

### Разделение на гармоническую и перкуссионную части

```python
# Seperation of Harmonic and Percussive Signals
y_harmonic, y_percussive = librosa.effects.hpss(y)
plt.figure(figsize=(15, 5))
librosa.display.waveshow(y_harmonic, sr=sr, alpha=0.25)
librosa.display.waveshow(y_percussive, sr=sr, color='r', alpha=0.5)
plt.title('Harmonic + Percussive')
```

## Выделение признаков

Извлечение признаков из аудиофайла
У сэмпла есть несколько важных признаков. Есть фундаментальное понятие ритма в некоторых формах, а остальные либо имеют свою нюансы, либо связаныритмом.

### Темп

Темп: скорость, с которой паттерны повторяются. Темп измеряется в битах в минуту (BPM). Если у музыки 120 BPM, это значит, что каждую минуту в ней 120 битов (ударов).
Бит: отрезок времени. Это ритм, выстукиваемый в песне. Так, в одном такте 4 бита, например.
Такт: логичное деление битов. Обычно в такте 3 или 4 бита, хотя возможны и другие варианты.
Интервал: в программах для редактирования чаще всего встречаются интервалы. Обычно есть последовательность нот, например, 8 шестнадцатых одинаковой длины. Обычно интервал — 8 нот, триплеты или четверные.
Ритм: список музыкальных звуков. Все ноты и являются ритмом.

#### В чем была ошибка?

Ошибка была в том что в новых версиях библиотеки <code>Seaborn</code> функция <code>barplot()</code> требует <code>keyword-only</code> аргументы, то есть <code>beat_nums</code> и <code>beat_time_diff</code> необходимо передавать как <code>x=beat_nums</code> и <code>y=beat_time_diff</code>, а также для того чтобы столбцы были раскрашены по градиенту необходимо также передать аргумент <code>hue=bet_nums</code>.

```python
# Beat Extraction
tempo, beat_frames = librosa.beat.beat_track(y=y_percussive,sr=sr)
print('Detected Tempo: '+str(tempo)+ ' beats/min')
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
beat_time_diff=np.ediff1d(beat_times)
beat_nums = np.arange(1, np.size(beat_times))

print(beat_nums)



fig, ax = plt.subplots()
fig.set_size_inches(6, 5)
ax.set_ylabel("Time difference (s)")
ax.set_xlabel("Beats")
g=sns.barplot(x=beat_nums, y=beat_time_diff,  palette="BuGn_d",ax=ax, hue=beat_nums, legend=False)
g=g.set(xticklabels=[])
```

### Мел-кепстральные коэффициенты (MFCC)

```python
# Calculate MFCCs
mfccs = librosa.feature.mfcc(y=y_harmonic, sr=sr, n_mfcc=20)

plt.figure(figsize=(15, 5))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
```

### Спектральный центроид - Spectral Centroid.

Указывает, на какой частоте сосредоточена энергия спектра или, другими словами, указывает, где расположен “центр масс” для звука. 
librosa.feature.spectral_centroid вычисляет спектральный центроид для каждого фрейма в сигнале:

```python
# Spectral Centroid
cent = librosa.feature.spectral_centroid(y=y, sr=sr)
plt.figure(figsize=(15,5))
plt.subplot(1, 1, 1)
plt.semilogy(cent.T, label='Spectral centroid')
plt.ylabel('Hz')
plt.xticks([])
plt.xlim([0, cent.shape[-1]])
plt.legend()
```

```python
import sklearn
spectral_centroids = cent[0]
spectral_centroids.shape

# Вычисление временной переменной для визуализации
plt.figure(figsize=(12, 4))
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
# Нормализация спектрального центроида для визуализации
def normalize(y, axis=0):
    return sklearn.preprocessing.minmax_scale(y, axis=axis)
# Построение спектрального центроида вместе с формой волны
librosa.display.waveshow(y, sr=sr, alpha=0.4)
plt.plot(t, normalize(spectral_centroids), color='b')
```

## Что такое гармоническая и перкуссионная часть сигнала, зачем они нужны?

Гармоническая часть сигнала содержит устойчивые звуки, которые обычно ассоциируются с музыкальными инструментами (гитара, скрипка и т.д.). Это звуки с определенной высотой (тонами) и частотами, которые создают мелодию и аккорды. Гармонические звуки имеют более длительные ноты и плавные изменения в амплитуде.



```python
IPython.display.Image(url="https://ok-english.ru/wp-content/uploads/2014/06/string-550x412.jpg", width=400, height=300)
```

Перкуссионная часть сигнала это часть сигнала которая включает в себя короткие, резкие звуки, которые часто ассоциируются с ударными инструментами (барабаны, щетки и т.д.) Эти звуки обычно не имеют четкой высоты и характеризуются быстрыми изменениями в амплитуде. Перкуссия добавляет ритм и текстуру в музыкальные произведения.

### Зачем это нужно?

Сепарация гармонической и перкуссионной частей сигнала позволяет более детально анализировать и обрабатывать аудиосигналы. Это может быть полезно в различных областях:
+ Музыкальная информация ( извлечение мелодий, аккордов и ритмов для анализа музыкальных произведений.)
+ Машинное обучение (подготовка данных для обучения моделей, которые могут распознавать или генерировать музыку.)
+ Аудиоэффекты ( применение различных эффектов только к определенным частям сигнала, например, добавление реверберации только к гармоническим звукам.)
